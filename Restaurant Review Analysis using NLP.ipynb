## Data Processing

```python
import nltk
import re
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
stop = stopwords.words('english')
```

```python
cleaned_review = []

for i in range(len(data)):
    review = re.sub('[^a-zA-Z]', ' ', data['Review'][i])    
    review = review.lower()    
    review = review.split()    
    review = [ps.stem(word) for word in review if word not in stop]    
    review = " ".join(review)    
    cleaned_review.append(review)
#print(cleaned_review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500, ngram_range = (1,2))
X = cv.fit_transform(cleaned_review).toarray()
y = data.iloc[:,1].values # df['Liked']
```

```python
cv.get_feature_names() # Use of n_grams in the range of 1, 2 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

# Building Models and Predicting Accuracy

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.naive_bayes import GaussianNB
clf1 = GaussianNB()
clf1.fit(X_train, y_train)
y_pred1 = clf1.predict(X_test)
print('Accuracy for Gaussian Naive Bayes: ', accuracy_score(y_test, y_pred1)*100)
print()
print(confusion_matrix(y_test, y_pred1))
print()
print("Classification Report for Gaussian Naive Bayes: ", classification_report(y_test, y_pred1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf2 = KNeighborsClassifier()
clf2.fit(X_train, y_train)
y_pred2 = clf2.predict(X_test)
accuracy_score(y_test, y_pred2)
print('Accuracy for KNN: ', accuracy_score(y_test, y_pred2)*100)
print()
print(confusion_matrix(y_test, y_pred2))
print()
print("Classification Report for KNN: ", classification_report(y_test, y_pred2))
```

```python
from sklearn.linear_model import LogisticRegression
clf3 = LogisticRegression()
clf3.fit(X_train, y_train)
y_pred3 = clf3.predict(X_test)
print('Accuracy for Logistic Regression: ', accuracy_score(y_test, y_pred3)*100)
print()
print(confusion_matrix(y_test, y_pred3))
print()
print("Classification Report for Logistic Regression: ", classification_report(y_test, y_pred3))
```

```python
from sklearn.svm import SVC
clf4 = SVC()
clf4.fit(X_train, y_train)
y_pred4 = clf4.predict(X_test)
print('Accuracy for Support Vector Classifier: ', accuracy_score(y_test, y_pred4)*100)
print()
print(confusion_matrix(y_test, y_pred4))
print()
print("Classification Report for Support Vector Classifier: ", classification_report(y_test, y_pred4))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf5 = DecisionTreeClassifier()
clf5.fit(X_train, y_train)
y_pred5 = clf5.predict(X_test)
print('Accuracy for Decision Tree Classifier: ', accuracy_score(y_test, y_pred5)*100)
print()
print(confusion_matrix(y_test, y_pred5))
print()
print("Classification Report for Decision Tree Classifier: ", classification_report(y_test, y_pred5))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf6 = RandomForestClassifier()
clf6.fit(X_train, y_train)
y_pred6 = clf6.predict(X_test)
print('Accuracy for Random Forest Classifier: ', accuracy_score(y_test, y_pred6)*100)
print()
print(confusion_matrix(y_test, y_pred6))
print()
print("Classification Report for Random Forest Classifier: ", classification_report(y_test, y_pred6))
```

By the use of Different Classifiers, we can say that the **Support Vector Classifier**  is the best for Restaurant Review Analysis with **75.50%** Accuracy.
